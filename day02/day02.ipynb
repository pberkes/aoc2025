# Part 1

```python
def parse_input(txt):
    return [tuple(int(x) for x in interval.split('-')) for interval in txt.split(',')]
```

```python
def is_valid(id_):
    str_id = str(id_)

    # Odd number of digits
    if len(str_id) % 2 == 1:
        return True

    half = len(str_id) // 2
    return str_id[:half] != str_id[half:]


test_cases = [
    (1188511885, False),
    (1698522, True),
]

for id_, expected in test_cases:
    print(id_)
    result = is_valid(id_)
    assert result == expected
```

```python
intervals = parse_input(txt)
intervals

tot = 0
for interval in intervals:
    for x in range(interval[0], interval[1]):
        if not is_valid(x):
            tot += x
tot
```

```python
def is_valid(id_):
    str_id = str(id_)

    len_id = len(str_id)
    for i in range(1, len_id // 2 + 1):
        # Length not divisible by `i`, can't be a repetition
        if len_id % i != 0:
            continue
        if str_id == str_id[:i] * (len_id // i):
            return False
    return True


test_cases = [
    (1188511885, False),
    (1698522, True),
    (999, False),
    (2121212121, False),
]

for id_, expected in test_cases:
    result = is_valid(id_)
    print(id_, result == expected)

999
```
